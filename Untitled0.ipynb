```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def data_to_array(file, columns, names=None):
    with open(file, 'br') as f:
        first_line = f.readline().decode().strip().split(';')
        print(first_line)
        list_columns = []
        for column in columns:
            list_columns.append(first_line.index(column))

        if names is True:
            names=first_line

        converters = {i: lambda s: float(s.decode().replace(',', '.')) for i in list_columns}

        A = np.genfromtxt(f, delimiter=';', names=names, usecols=list_columns,
                          converters=converters)

        return A
```

```python
avec = data_to_array("26-11/avec plaque vrai.csv", ('Temps2', 'EA2', 'EA3'), names=True)
sans = data_to_array("26-11/avec plaque.csv", ('Temps2', 'EA2', 'EA3'), names=True)

plt.subplot(211)
plt.plot(avec['Temps2'], avec['EA2'])
plt.twinx()
plt.plot(avec['Temps2'], avec['EA3'], 'r')
plt.title('Avec plaque')

plt.subplot(212)
plt.plot(sans['Temps2'], sans['EA2'])
plt.twinx()
plt.plot(sans['Temps2'], sans['EA3'], 'r')
plt.title('Sans plaque')
```

```python
from matplotlib.widgets import Cursor

fig = plt.figure()

ax = fig.add_subplot(111)

plt.plot(sans['Temps2'][:10000], - avec['EA3'][:10000] + sans['EA3'][:10000], 'b')
plt.title('Différence') 

cursor = Cursor(ax)
```
